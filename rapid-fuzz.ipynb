```python
from rapidfuzz import fuzz, process
from fuzzywuzzy import fuzz as fuzz2
import pandas as pd
import random, string, time
```

```python
def gen_random_string(n):
  '''
  input: n, desired length of random string
  output: a randomly generated string of length n
  '''
  letters = string.ascii_lowercase
  return ''.join(random.choice(letters) for i in range(n))
```

```python
ipn = pd.read_csv('IPN.csv', usecols = ['OWNERS'])
ipn.head(5)
```

```python
len(ipn)
```

```python
smes = pd.read_csv('smes.csv', usecols = ['Company'])
smes.head(5)
```

```python
len(smes)
```

```python
ipn.dropna(inplace=True) # Drop nulled values.

start = time.time()
smes['matched'] = smes['Company'].apply(lambda x: process.extractOne(x, ipn['OWNERS'].to_list(), score_cutoff = 90))
end = time.time()
print('It took {} seconds'.format(int(end - start)))
```

```python
smes[smes['matched'].notna()] # show values that aren't none.
```

## Testing Rapid Fuzz

```python
start = time.time()
for i in range(1000000):
  fuzz.ratio(gen_random_string(15), gen_random_string(15))
end = time.time()

print('It took {} seconds'.format(int(end - start)))
```

#Testing Fuzzy Wuzzy

```python
start = time.time()
for i in range(1000000):
  fuzz2.ratio(gen_random_string(15), gen_random_string(15))
end = time.time()

print('It took {} seconds'.format(int(end - start)))
```
